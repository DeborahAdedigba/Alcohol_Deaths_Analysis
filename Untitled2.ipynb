```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your alcohol data
alcohol_data = pd.read_csv("Cleaned_1_data.csv")  # Replace "your_alcohol_data.csv" with your actual file path

# Extract states and the corresponding number of deaths
states = alcohol_data['LocationDesc'].values
deaths = alcohol_data['Data_Value_Alt'].values.astype(float)  # Convert to float
deaths = deaths.reshape(-1, 1)  # Reshape to ensure it's a 2D array

# Combine states and number of deaths into one array
data = np.hstack((states.reshape(-1, 1), deaths))

class KMeans:
    def __init__(self, data, k, steps):
        self.data = data
        self.k = k
        self.steps = steps
        self.centers = self.data[np.random.choice(range(len(self.data)), self.k, replace=False)]
        self.colors = np.array(np.random.randint(0, 255, size=(self.k, 4))) / 255
        self.colors[:, 3] = 1

    def distance(self, p1, p2):
        return np.sqrt(np.sum((p1 - p2) ** 2))

    def initialize(self):
        """
        Initialize the Clusters.
        """
        self.clusters = {i: [] for i in range(self.k)}
        plt.scatter(range(len(self.data)), self.data[:, 1], s=100, color='blue')
        plt.title("Initialize")
        plt.show()

    def fit(self):
        for step in range(self.steps):
            self.clusters = {i: [] for i in range(self.k)}

            for point in self.data:
                d = np.array([self.distance(point[1:], c[1:]) for c in self.centers])  # Update indexing here
                c = np.argmin(d)
                self.clusters[c].append(point)

            self.clusters = {i: np.array(v) for i, v in self.clusters.items()}
            self.centers = np.array([np.mean(self.clusters[i], axis=0) for i in range(self.k)])
            self.visualize(step=step)

    def visualize(self, step):
        plt.title(f"Step : {step}")
        [plt.scatter(range(len(self.clusters[i])), self.clusters[i][:, 1], marker="*", s=100,
                     color=self.colors[i]) for i in range(self.k)]
        plt.plot(range(len(self.centers)), self.centers[:, 1], 'ro-', markersize=10)  # Update indexing here
        plt.show()

# Number of clusters and steps
k = 3  # Number of clusters
steps = 5  # Number of steps

# Initialize and fit KMeans
kmeans = KMeans(data, k, steps)
kmeans.initialize()
kmeans.fit()
```
